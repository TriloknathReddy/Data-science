1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
nums = [10, 20, 30, 40, 50]
print('Original list:', nums)

# append
nums.append(60)
print('After append(60):', nums)

# extend
nums.extend([70, 80])
print('After extend([70, 80]):', nums)

# insert
nums.insert(1, 15)
print('After insert(1, 15):', nums)

# remove
nums.remove(30)
print('After remove(30):', nums)

# pop
popped_value = nums.pop()
print('After pop():', nums)
print('Popped value:', popped_value)

# index
sample_list = [10, 20, 30, 40, 50]
index_30 = sample_list.index(30)
print('Index of 30 in [10, 20, 30, 40, 50]:', index_30)

# count
count_10 = [10, 10, 20].count(10)
print('Count of 10 in [10, 10, 20]:', count_10)

# sort and reverse
arr = [5, 2, 9, 1]
print('Before sort:', arr)
arr.sort()
print('After sort():', arr)
arr.reverse()
print('After reverse():', arr)
```

5.	 Write a script that checks if a person is eligible to vote (age ≥ 18). Take age as a variable and print the appropriate message.

```python

# Check if a person is eligible to vote age >= 18

def check_voting_eligibility(age):
    'Prints whether the person is eligible to vote based on age.'
    if age >= 18:
        print('Age:', age, '- Eligible to vote')
    else:
        print('Age:', age, '- Not eligible to vote')

# Example
age = 17
check_voting_eligibility(age)
```

6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'


```python
# Print grade according to marks.

def get_grade(marks):
    'Returns the grade based on marks.'
    if marks >= 90:
        return 'A'
    elif marks >= 75:
        return 'B'
    elif marks >= 50:
        return 'C'
    else:
        return 'Fail'

# Example
marks = 87
print('Marks:', marks, 'Grade:', get_grade(marks))
```

7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
def check_number(num):
    'Checks if a number is positive and even, or zero, or negative.'
    if num > 0:
        if num % 2 == 0:
            print(num, 'is positive and even')
        else:
            print(num, 'is positive but odd')
    elif num == 0:
        print('The number is zero')
    else:
        print(num, 'is negative')

# Example
number = -5
check_number(number)
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np

data = {
    'Name': ['A','B','C','D','E','F','G','H','I','J'],
    'Age': [20, 22, 21, 23, 20, 22, 23, 21, 22, 20],
    'City': ['BLR', 'BLR', 'HYD', 'BLR', 'HYD', 'BLR', 'DEL', 'BLR', 'HYD', 'DEL'],
    'Salary': [30000, 25000, 27000, None, 32000, 29000, None, 31000, 33000, 28000]
}

df = pd.DataFrame(data)
print('Original DataFrame:')
print(df)

# .info()
print('\nDataFrame info:')
print(df.info())

# .describe()
print('\nDataFrame description:')
print(df.describe())

# Select Name and City columns
print('\nSelected columns (Name and City)')
print(df[['Name', 'City']])

# Drop City column
df_without_city = df.drop('City', axis=1)
print("\nDataFrame after dropping ,'City' column:")
print(df_without_city)

# Fill missing values in Salary with mean
mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print('\nDataFrame after filling missing Salary values with mean:')
print(df)

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print('\nDataFrame after removing duplicates:')
print(df_no_duplicates)
```
