### Exercise 1 Prime Numbers

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True
number = int(input('Enter a number: '))
if is_prime(number):
    print(number, 'is a prime number.')
else:
    print(number, 'is not a prime number.')
```

### Exercise 2 Product of Random Numbers

```python
import random

num1 = random.randint(1, 100)
num2 = random.randint(1, 100)

correct = num1 * num2
answer = int(input(f'What is the product of {num1} times {num2}? '))

if answer == correct:
    print('Correct!')
else:
    print(f'Wrong! The correct answer is {correct}')
```

### Exercise 4 Problem Statement Hospital Billing System


```python
# Charges
roomcharge_perday = 2000
doctor_fee = 1500
lab_test_charges = 300
medicine_charges = 2400

days_stayed = 4
lab_tests = 3

room_total = roomcharge_perday * days_stayed
lab_total = lab_test_charges * lab_tests

total_before_discount = room_total + doctor_fee + lab_total + medicine_charges

if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

final_bill = total_before_discount - discount

print('HOSPITAL BILL')
print('Room Charges (₹2000 x', days_stayed, "days): ₹", room_total)
print('Doctor Consultation Fee: ₹', doctor_fee)
print('Lab Test Charges (₹300 x, lab_tests, "tests): ₹', lab_total)
print('Medicine Charges: ₹', medicine_charges)
print('Total before discount: ₹', total_before_discount)
print('Discount: ₹', discount)
print('Final Bill: ₹', final_bill)
```
